```python
#Binning of horsepower column 
#created a bin array using numpy linespace function
bins = np.linspace(min(Auto_df["horsepower"]), max(Auto_df["horsepower"]), 4)
bins
```

```python
#setting the group names
group_names = ['Low', 'Medium', 'High']
```

```python
#use cut function to determine where each value of horsepower belongs to
Auto_df['horsepower-binned'] = pd.cut(Auto_df['horsepower'], bins, labels=group_names,include_lowest=True )
Auto_df[['horsepower','horsepower-binned']].head(20)
```

```python
#histogram to visualize the distribution of bins
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(Auto_df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
